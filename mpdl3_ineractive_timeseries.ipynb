```python
import numpy as np
import matplotlib.pyplot as plt
import mpld3
from mpld3 import plugins, fig_to_html
import pandas as pd
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
hob = pd.read_csv('H26_20190531.csv', skiprows=2, usecols=[1, 2, 3],
                              names=['time', 'T', 'RH'],
                              index_col='time', parse_dates=True).resample('W').mean()
```

```python
css = """
table
{
  border-collapse: collapse;
}

td
{
  background-color: #cccccc;
}
table,  td
{
  font-family:Arial, Helvetica, sans-serif;
  border: 1px #1f77b4;
  text-align: left;
}
"""
css = """style="background-color: #FFFF00"""

fig, ax = plt.subplots()
ax.grid(alpha=0.3)
points = ax.plot(hob.index,hob['T'],marker='o')

labels = []
for i in range(len(hob)):
    value = hob.iloc[i]['T']
    value = f"{value:.3f}"
    label = f"<p style='background: #1f77b4'>{value}</p>"
    labels.append(label)

tooltip = plugins.PointHTMLTooltip(points[0], labels,
                                   voffset=10, hoffset=10, css=css)
plugins.connect(fig, tooltip)
mpld3.display()
# mpld3.show()
```
